```r
plot(pca_result$rotation[,1], pca_result$rotation[,2], col = "white",
     xlim = c(-.35, .1), ylim = c(-.4, .3),
     xlab = "PC1", ylab = "PC2")
text(pca_result$rotation[,1], pca_result$rotation[,2], rownames(pca_result$rotation))
```

```r
plot(pca_result$x[,1],pca_result$x[,2],col= "white",
     xlim = c(-3.75, 3.75),ylim= c(-3.75,3.75),
     xlab = "PC1",ylab = "PC2")
text(
  pca_result$x[,1], pca_result$x[,2],
  breast_cancer_data$Diagnosis
)
```

```r
library(ggplot2)

var_explained = pca_result$sdev^2 / sum(pca_result$sdev^2)
ggplot(data = data.frame(x = c(1:10), y = var_explained),
       mapping = aes(x = x, y = y)) + geom_line() +
  xlab("PCs") + ylab("Variance Explained") + ggtitle("Breast Cancer Dataset Plot ") +
  ylim(0, 1)
```

```r
install.packages("Rtsne")
library(Rtsne)
```

```r
features <- breast_cancer_data[, -(1:2)]
labels <- breast_cancer_data$Diagnosis
scaled_features <- scale(features)
set.seed(42)
bc_tsne_result <- Rtsne(scaled_features)
plot(bc_tsne_result$Y, col = as.factor(labels), pch = 19, main = "t-SNE Visulaization of Breast Cancer Dataset")
legend("bottomleft", legend = unique(labels), col = 1:length(unique(labels)), pch = 19, title = "Diagnosis")
```

```r
features <- breast_cancer_data[, -(1:2)]
labels <- breast_cancer_data$Diagnosis
scaled_features <- scale(features)
set.seed(42)
bc_tsne_result <- Rtsne(scaled_features, perplexity = 8)
plot(bc_tsne_result$Y, col = as.factor(labels), pch = 19, main = "t-SNE Visulaization of Breast Cancer Dataset")
legend("bottomleft", legend = unique(labels), col = 1:length(unique(labels)), pch = 19, title = "Diagnosis")
```
